
$ s(\vec x)= i(\vec x ')+ R_{tip} \left[1 - \left(1 + \left( \frac{di}{d\vec x'}\right)^2 \right)^{-\frac{1}{2}} \right] $

$ R_i = R_s + R_t $

$ x_c ^2 + y_c ^2 = (R_s + R_t)^2 $

\begin{align}
    y &= \begin{bmatrix}
           x_c \\
           y_c - R_t \\
         \end{bmatrix}
  \end{align}
  
\begin{align}
    y(x_c) &= \begin{bmatrix}
           x_c \\
           \sqrt{(R_s + R_t)^2 - x_c^2} - R_t \\
         \end{bmatrix}
  \end{align}

$ \frac{N^2(N^2 -1)}{2} $

```python
import numpy as np
import matplotlib.pyplot as plt

R_p=10
R_t=10
x_c=np.linspace(-R_p, R_p, 20)

a=4*R_p**2 + x_c**2
b=2*x_c**2 *(x_c**2+3*R_p**2-R_t**2)
c=x_c**2 *(x_c**4 + R_p**4 + R_t**4 +2*x_c**2 * R_p**2 -2*x_c**2 * R_t**2 -2*R_p**2 * R_t**2)

print(b**2 -4*a*c)
y_c=(-b +np.sqrt(b**2 -4*a*c))/(2*a)

plt.figure(figsize=(10,10))
plt.plot(x_c,y_c)
#plt.title('Optical correction: m='+str(round(m.real*n_wat, 3) + round(m.imag*n_wat, 3) * 1j)+', $\lambda$='+str(round(lambda0*n_wat*10**9, 1))+'nm')
#plt.xlabel("Scattering angle (degrees)")
#plt.ylabel("Intensity (1/sr)")
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

R_p=1
R_t=2
nx=150
boundary=np.sqrt((R_p+R_t)**2-R_t**2)+0.5
x_c=np.linspace(-boundary, boundary, nx)

plt.figure(figsize=(8,4))
plt.plot(x_c,np.sqrt((R_p+R_t)**2-x_c**2) -R_t, label='analytic')
plt.plot(x_c,np.sqrt(R_p**2-x_c**2), label='particle')
cut=np.arange(16)
x_c=np.delete(x_c, np.concatenate((cut,nx-cut)))
plt.plot(x_c,np.sqrt((R_p+R_t)**2-x_c**2) -R_t, c='red', label='dilation')
plt.title(r'$ R_{part}='+str(R_p)+r'; \quad R_{tip}='+str(2)+r'$')
plt.xlabel('x')
plt.ylabel('z')
plt.xlim(-boundary, boundary)
plt.ylim(-boundary/2, boundary/2)
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

R_p=1
R_t=2
nx=150
cx=-0.8
cy=np.sqrt((R_t+R_p)**2 - cx**2)
boundary=np.sqrt((R_p+R_t)**2-R_t**2)
x_d=np.linspace(-boundary, cx, nx)
x_p=np.linspace(-R_p, R_p, nx)
x_t=np.linspace(-R_t, R_t, nx)+cx

plt.figure(figsize=(8,8))
plt.plot([cx,0],[cy,0])
plt.plot(x_t,-np.sqrt(R_t**2-(x_t-cx)**2)+cy, c='blue', label='tip')
#plt.plot(x_t, np.sqrt(R_t**2-(x_t-cx)**2)+c_t[1], c='blue')

plt.plot(x_c,np.sqrt(R_p**2-x_c**2), label='particle')
plt.plot(x_d,np.sqrt((R_p+R_t)**2-x_d**2) -R_t, c='red', label='dilation')
plt.scatter(cx,cy-R_t, marker='*', c='red')
#plt.title(r'$ R_{part}='+str(R_p)+r'; \quad R_{tip}='+str(2)+r'$')
#plt.xlabel('x')
#plt.ylabel('z')
plt.xlim(-3, 1.2)
plt.ylim(-0.6, 3.6)
plt.legend()
plt.grid(True)
plt.show()
```
